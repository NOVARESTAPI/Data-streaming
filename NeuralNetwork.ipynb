```python
# -------------------------------------------------------------------------------------
# Full Ternary Label ML Pipeline + Backtest
# -------------------------------------------------------------------------------------

import numpy as np
import pandas as pd
import yfinance as yf
import pandas_ta as ta
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import regularizers
from keras.callbacks import EarlyStopping

# -------------------------------------------------------------------------------------
# Step 1: Download Data
# -------------------------------------------------------------------------------------

ticker = "O39.SI"
data = yf.download(ticker, start="2008-01-01", end="2026-01-01")
data.columns = data.columns.droplevel(level=1)  # remove multi-level columns if present

# -------------------------------------------------------------------------------------
# Step 2: Feature Engineering
# -------------------------------------------------------------------------------------

data["RET_1d"] = data["Close"].pct_change(1)
data["RET_5d"] = data["Close"].pct_change(5)
data["RET_10d"] = data["Close"].pct_change(10)

data["SMA_5"] = data["Close"].rolling(5).mean()
data["SMA_20"] = data["Close"].rolling(20).mean()
data["SMA_ratio"] = data["SMA_5"] / data["SMA_20"]

bb = ta.bbands(data["Close"], length=20, std=2)

# Dynamically find Bollinger Band column names
if bb is not None:
    bb_upper_cols = bb.filter(like='BBU')
    bb_lower_cols = bb.filter(like='BBL')

    if not bb_upper_cols.empty and not bb_lower_cols.empty:
        bb_upper = bb_upper_cols.columns[0]
        bb_lower = bb_lower_cols.columns[0]
        data["BB_width"] = (bb[bb_upper] - bb[bb_lower]) / data["SMA_20"]
    else:
        print("Warning: Could not find Bollinger Band columns.")
        data["BB_width"] = np.nan
else:
    print("Warning: Bollinger Bands could not be calculated.")
    data["BB_width"] = np.nan


data["RSI"] = ta.rsi(data["Close"], length=14)
macd = ta.macd(data["Close"])
if macd is not None and "MACD_12_26_9" in macd.columns:
    data["MACD"] = macd["MACD_12_26_9"]
else:
    print("Warning: MACD could not be calculated.")
    data["MACD"] = np.nan

adx = ta.adx(data["High"], data["Low"], data["Close"], length=14)
if adx is not None and "ADX_14" in adx.columns:
    data["ADX"] = adx["ADX_14"]
else:
    print("Warning: ADX could not be calculated.")
    data["ADX"] = np.nan


data["VOLATILITY"] = data["RET_1d"].rolling(14).std()
data["VOL_MA"] = data["Volume"].rolling(20).mean()
data["VOL_ratio"] = data["Volume"] / data["VOL_MA"]


# -------------------------------------------------------------------------------------
# Step 3: Define Ternary Labels (5-day forward returns)
# -------------------------------------------------------------------------------------

horizon = 3 # 3 days forward
data["Fwd_Returns"] = data["Close"].pct_change(horizon).shift(-horizon)

# Define thresholds (e.g., top 30% and bottom 30% of returns)
up_thresh = data["Fwd_Returns"].quantile(0.7)
down_thresh = data["Fwd_Returns"].quantile(0.3)

def label_return_ternary(r):
    if r > up_thresh:
        return 1   # Up
    elif r < down_thresh:
        return 0  # Down
    else:
        return -1 # Neutral

data["Signal_ternary"] = data["Fwd_Returns"].apply(label_return_ternary)

# -------------------------------------------------------------------------------------
# Step 4: Prepare Dataset (drop NA and neutral labels)
# -------------------------------------------------------------------------------------

features = [
    "RET_1d","RET_5d","RET_10d",
    "SMA_ratio","BB_width","RSI","MACD","ADX",
    "VOLATILITY","VOL_ratio"
]

df = data.dropna().copy() # Use .copy() to avoid SettingWithCopyWarning
df = df[df["Signal_ternary"] != -1].copy() # Drop neutral labels and use .copy()

X = df[features]
y = df["Signal_ternary"] # Use ternary signal

# Convert labels to one-hot encoding for Keras
y = pd.get_dummies(y, prefix="Signal").values

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# -------------------------------------------------------------------------------------
# Step 5: Define Model
# -------------------------------------------------------------------------------------

model = Sequential()
model.add(Dense(64, activation="relu", input_dim=X_train.shape[1],
                kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.3))
model.add(Dense(32, activation="relu",
                kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.3))
model.add(Dense(y_train.shape[1], activation="softmax")) # Softmax for ternary classification

model.compile(optimizer="adam",
              loss="categorical_crossentropy", # Categorical crossentropy for ternary classification
              metrics=["accuracy"])

# -------------------------------------------------------------------------------------
# Step 6: Train Model
# -------------------------------------------------------------------------------------

early_stop = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)

history = model.fit(X_train, y_train,
                    epochs=200,
                    batch_size=32,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stop])

# -------------------------------------------------------------------------------------
# Step 7: Training Curves
# -------------------------------------------------------------------------------------

plt.figure(figsize=(10,5))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curves")
plt.legend()
plt.show()

plt.figure(figsize=(10,5))
plt.plot(history.history["accuracy"], label="Train Acc")
plt.plot(history.history["val_accuracy"], label="Val Acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Accuracy Curves")
plt.legend()
plt.show()

# -------------------------------------------------------------------------------------
# Step 8: Evaluation on Test Set
# -------------------------------------------------------------------------------------

y_pred_prob = model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1) # Get predicted class index

# Convert y_test back from one-hot for confusion matrix and classification report
y_test_labels = np.argmax(y_test, axis=1)

print("Confusion Matrix:")
print(confusion_matrix(y_test_labels, y_pred))
print("\nClassification Report:")
print(classification_report(y_test_labels, y_pred))

# -------------------------------------------------------------------------------------
# Step 9: Backtest with Full Dataset
# -------------------------------------------------------------------------------------

# Use all data (including neutral) for predictions
df_all = data.dropna().copy() # Use .copy() to avoid SettingWithCopyWarning
X_all = sc.transform(df_all[features])
df_all["pred_prob"] = list(model.predict(X_all)) # Store probabilities as a list of arrays
df_all["pred_signal"] = np.argmax(model.predict(X_all), axis=1) # Get predicted class index

# Map predicted index back to original labels (0: Down, 1: Up)
# Note: This assumes the one-hot encoding order is consistent.
# If the order is not guaranteed, you might need to store the mapping.
label_mapping = {0: 0, 1: 1} # Assuming 0 is Down, 1 is Up after one-hot encoding
df_all["pred_signal_mapped"] = df_all["pred_signal"].map(label_mapping)


# Strategy: Long if pred_signal=1, Short if pred_signal=0, Neutral if pred_signal=-1 (or not in mapping)
# For simplicity in backtest, assuming we can short. In reality, this might need adjustment.
# Filter out neutral signals for strategy returns calculation
df_strategy = df_all[df_all["Signal_ternary"] != -1].copy() # Use .copy() to avoid SettingWithCopyWarning

df_strategy["strategy_returns"] = np.where(df_strategy["pred_signal_mapped"]==1,
                                      df_strategy["RET_1d"], -df_strategy["RET_1d"])


# Cumulative returns for strategy and buy & hold (using original data for buy & hold)
df_all["cum_buyhold"] = (1 + df_all["RET_1d"]).cumprod()

# -------------------------------------------------------------------------------------
# Step 10: Plot Backtest Results
# -------------------------------------------------------------------------------------

fig, ax1 = plt.subplots(figsize=(12, 6))
df_strategy["strategy_cum_returns"] = (1 + df_strategy["strategy_returns"]).cumprod()

ax1.plot(df_strategy.index, df_strategy["strategy_cum_returns"], label="Strategy", color='blue')
ax1.set_ylabel("Strategy Cumulative Returns", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(df_all.index, df_all["cum_buyhold"], label="Buy & Hold", linestyle="--", color='orange')
ax2.set_ylabel("Buy & Hold Cumulative Returns", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title(f"Backtest: {ticker}")
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))
plt.show()

# Show last 10 days of signals
print(df_all[["Close", "Signal_ternary", "pred_signal_mapped", "Fwd_Returns"]].tail(10))
```
